```python
import matplotlib.pyplot as plt
import numpy as np
```

# Joint Attack

## Queries vs Stepsize

```python
stepsizes = ['0.7 - SimBA', '0.8 - SimBA', '0.9 - SimBA', '0.1 - NES']
x = np.arange(len(stepsizes))
queries_by_hamm_thresh = {
    "Hamming Threshold = 1": np.array([1112, 964, 940, 139]),
    "Hamming Threshold = 5": np.array([1522, 1438, 1374, 205]),
    "Hamming Threshold = 10": np.array([3435, 3368, 3197, 261]),
    "Hamming Threshold = 19": np.array([4507, 4413, 4206, 437])
}
width = 0.20
multiplier = 0

fig, ax= plt.subplots(layout='constrained')
for attribute, measurement in queries_by_hamm_thresh.items():
    offset = width*multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1
fig.set_figwidth(8)
ax.set_ylabel('Queries')
ax.set_xlabel('Stepsize by Method')
ax.set_xticks(x+width, stepsizes)
ax.legend(loc='upper right')
ax.set_ylim(0, 5900)
plt.show()
```

## ASR vs Hamming Threshold

```python

hamming_thresholds = [1, 5, 10, 19]
x = np.arange(len(hamming_thresholds))
asr_by_stepsize = {
    "SimBA Stepsize = 0.7": np.array([100, 100, 99, 86]),
    "SimBA Stepsize = 0.8": np.array([100, 100, 98, 85]),
    "SimBA Stepsize = 0.9": np.array([100, 100, 98, 89]),
    "NES Stepsize = 0.1": np.array([100, 100, 100, 94])
}
width = 0.20
multiplier = 0

fig, ax = plt.subplots(layout='constrained')
for attribute, measurement in asr_by_stepsize.items():
    offset = width*multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1
ax.set_ylabel('Attack Success Rate (%)')
ax.set_xlabel('Hamming Threshold (/96)')
ax.set_xticks(x+width, hamming_thresholds)
ax.legend(loc='upper right')
ax.set_ylim(0, 140)
plt.show()
```
